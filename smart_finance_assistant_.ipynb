### Foundation Data Processing Functions

````python
# 🤖 AI Collaboration: Data Loading and Cleaning Function
# This function was developed from my pseudocode to clean transaction data,
# handle currency symbols, separate refunds, and generate budget insights.

def clean_amount_value(amount):
    """
    Converts messy amount values such as '$45.50', 'AUD 12.00', or '-$25.00'
    into numeric float values.
    """
    try:
        amount_text = str(amount)
        amount_text = amount_text.replace('$', '')
        amount_text = amount_text.replace('AUD', '')
        amount_text = amount_text.replace(',', '')
        amount_text = amount_text.strip()
        return float(amount_text)
    except:
        return None


def analyze_spending_data(transaction_data, weekly_budget=None):
    """
    Analyzes transaction data and returns cleaned data, category summaries,
    refund totals, net spending, and simple budget insights.
    """

    required_columns = ['Date', 'Amount', 'Category', 'Description']

    # Check required columns
    for column in required_columns:
        if column not in transaction_data.columns:
            return f"Error: Missing required column: {column}"

    # Copy data so the original dataset is not changed
    cleaned_data = transaction_data.copy()

    # Clean category values
    cleaned_data['Category'] = cleaned_data['Category'].fillna('Unknown')

    # Clean amount values
    cleaned_data['Clean Amount'] = cleaned_data['Amount'].apply(clean_amount_value)

    # Remove rows where amount cannot be converted
    cleaned_data = cleaned_data.dropna(subset=['Clean Amount'])

    if cleaned_data.empty:
        return "Error: No valid transaction data available for analysis."

    # Separate spending and refunds
    spending_data = cleaned_data[cleaned_data['Clean Amount'] > 0]
    refund_data = cleaned_data[cleaned_data['Clean Amount'] < 0]

    # Calculate totals
    total_positive_spending = spending_data['Clean Amount'].sum()
    total_refunds = abs(refund_data['Clean Amount'].sum())
    net_spending = total_positive_spending - total_refunds

    # Category summary
    category_summary = (
        spending_data
        .groupby('Category')['Clean Amount']
        .agg(['sum', 'count', 'mean'])
        .reset_index()
        .sort_values(by='sum', ascending=False)
    )

    category_summary.columns = ['Category', 'Total Spent', 'Transaction Count', 'Average Transaction']

    # Highest spending category
    if not category_summary.empty:
        highest_category = category_summary.iloc[0]['Category']
        highest_category_amount = category_summary.iloc[0]['Total Spent']
    else:
        highest_category = "None"
        highest_category_amount = 0

    # Budget comparison
    budget_result = "No weekly budget was provided."

    if weekly_budget is not None:
        try:
            weekly_budget = float(weekly_budget)

            if weekly_budget <= 0:
                budget_result = "Invalid budget: budget must be greater than zero."
            elif total_positive_spending > weekly_budget:
                over_budget = total_positive_spending - weekly_budget
                budget_result = f"You are AUD {over_budget:.2f} over your weekly budget."
            else:
                remaining_budget = weekly_budget - total_positive_spending
                budget_result = f"You are within budget. AUD {remaining_budget:.2f} remains."

        except:
            budget_result = "Invalid budget input. Budget comparison was skipped."

    # Plain-English insights
    insights = []

    insights.append(f"Total positive spending before refunds was AUD {total_positive_spending:.2f}.")
    insights.append(f"Total refunds were AUD {total_refunds:.2f}.")
    insights.append(f"Net spending after refunds was AUD {net_spending:.2f}.")
    insights.append(f"The highest spending category was {highest_category}, with AUD {highest_category_amount:.2f} spent.")
    insights.append(budget_result)

    if highest_category in ['Entertainment', 'Coffee', 'Dining', 'Shopping']:
        insights.append(f"{highest_category} appears to be a discretionary spending area that may be worth reviewing.")

    return {
        'cleaned_data': cleaned_data,
        'category_summary': category_summary,
        'total_positive_spending': total_positive_spending,
        'total_refunds': total_refunds,
        'net_spending': net_spending,
        'budget_result': budget_result,
        'insights': insights
    }
````

````python
# Create sample transaction data for testing
import pandas as pd

sample_transactions = {
    'Date': ['2024-08-01', '2024-08-02', '2024-08-03', '2024-08-04', '2024-08-05'],
    'Amount': ['$45.50', '$12.00', '$89.95', '$3.50', '-$25.00'],
    'Category': ['Groceries', 'Transport', 'Entertainment', 'Coffee', 'Refund'],
    'Description': ['Woolworths', 'Opal Card', 'Concert Tickets', 'Campus Cafe', 'Returned Item']
}

df_sample = pd.DataFrame(sample_transactions)

# Run the analysis function
analysis_result = analyze_spending_data(df_sample, weekly_budget=120)

print("✅ Spending analysis completed successfully!")
print()

print("Total positive spending:", f"AUD {analysis_result['total_positive_spending']:.2f}")
print("Total refunds:", f"AUD {analysis_result['total_refunds']:.2f}")
print("Net spending:", f"AUD {analysis_result['net_spending']:.2f}")
print("Budget result:", analysis_result['budget_result'])

print("\nCategory Summary:")
print(analysis_result['category_summary'])

print("\nInsights:")
for insight in analysis_result['insights']:
    print("-", insight)
````

````python
# 🤖 AI Collaboration: Business Insights Generator
# This function uses the analysis results to create clear, practical,
# business-friendly financial recommendations for the user.

def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations based on spending analysis.

    Args:
        analysis_data: Dictionary containing spending analysis results.

    Returns:
        str: Formatted recommendations report.
    """

    total_spending = analysis_data['total_positive_spending']
    total_refunds = analysis_data['total_refunds']
    net_spending = analysis_data['net_spending']
    budget_result = analysis_data['budget_result']
    category_summary = analysis_data['category_summary']
    insights = analysis_data['insights']

    report = "SMART FINANCE ASSISTANT RECOMMENDATIONS\n"
    report += "=" * 48 + "\n\n"

    report += "1. Spending Overview\n"
    report += f"- Total positive spending before refunds: AUD {total_spending:.2f}\n"
    report += f"- Total refunds: AUD {total_refunds:.2f}\n"
    report += f"- Net spending after refunds: AUD {net_spending:.2f}\n"
    report += f"- Budget status: {budget_result}\n\n"

    report += "2. Category Spending Summary\n"

    if category_summary.empty:
        report += "- No category spending data was available.\n\n"
    else:
        for _, row in category_summary.iterrows():
            report += (
                f"- {row['Category']}: AUD {row['Total Spent']:.2f} "
                f"across {int(row['Transaction Count'])} transaction(s), "
                f"average AUD {row['Average Transaction']:.2f}\n"
            )
        report += "\n"

    report += "3. Key Insights\n"
    for insight in insights:
        report += f"- {insight}\n"

    report += "\n4. Recommendations\n"

    if not category_summary.empty:
        highest_category = category_summary.iloc[0]['Category']
        highest_amount = category_summary.iloc[0]['Total Spent']

        report += (
            f"- Review {highest_category}, because it is currently the highest "
            f"spending category at AUD {highest_amount:.2f}.\n"
        )

        if highest_category in ['Entertainment', 'Coffee', 'Dining', 'Shopping']:
            report += (
                f"- Since {highest_category} is usually discretionary, the user could "
                "set a small weekly limit for this category first.\n"
            )

    if "over your weekly budget" in budget_result:
        report += (
            "- The user is currently over budget, so the assistant should recommend "
            "reducing flexible spending before cutting essential expenses.\n"
        )
    elif "within budget" in budget_result:
        report += (
            "- The user is within budget, so the assistant could encourage saving "
            "the remaining amount or tracking the same pattern next week.\n"
        )

    if total_refunds > 0:
        report += (
            "- Refunds should continue to be tracked separately because they reduce "
            "net spending but should not be treated as normal spending.\n"
        )

    report += (
        "- A useful next step would be to add more transactions over a longer time "
        "period so the assistant can identify stronger spending patterns.\n"
    )

    return report


# Test the recommendation function using the analysis result created above
recommendations = generate_financial_recommendations(analysis_result)

print(recommendations)
````

## RAG System for Financial Documents

````python
# 🤖 AI Collaboration: RAG System using hands-on-ai
# Upgraded from a manual Python dictionary to the hands-on-ai RAG module.
# The RAG pattern retrieves relevant guidance before generating a response.
# Includes a graceful fallback if the server is unavailable.

financial_knowledge_base = [
    "A good budgeting strategy is to compare total spending against a weekly or monthly limit and review categories that exceed expectations.",
    "Refunds should be separated from normal spending because they reduce net spending but do not represent new purchases.",
    "Discretionary spending includes flexible categories such as entertainment, coffee, shopping, and dining. These are useful areas to review when a user is over budget.",
    "Savings recommendations should be realistic, specific, and linked to the user's actual spending pattern.",
    "The 50/30/20 rule suggests spending 50% of income on needs, 30% on wants, and saving 20%.",
    "Tracking spending by category helps identify patterns and makes it easier to find areas to reduce."
]

try:
    from hands_on_ai.rag import create_rag, retrieve

    rag_index = create_rag(financial_knowledge_base)

    def retrieve_financial_guidance(user_question):
        """Retrieves relevant finance guidance using hands-on-ai RAG."""
        results = retrieve(user_question, rag_index, top_k=2)
        return results if results else [
            "General guidance: review total spending, refunds, highest categories, and budget status."
        ]

    print("✅ RAG system using hands-on-ai created successfully.")

except Exception as e:
    print(f"ℹ️ hands-on-ai RAG unavailable ({e}). Using keyword fallback.")

    knowledge_dict = {
        "budgeting": financial_knowledge_base[0],
        "refunds": financial_knowledge_base[1],
        "discretionary": financial_knowledge_base[2],
        "savings": financial_knowledge_base[3],
        "50/30/20": financial_knowledge_base[4],
        "category": financial_knowledge_base[5]
    }

    def retrieve_financial_guidance(user_question):
        """Keyword-based fallback retrieval."""
        question = user_question.lower()
        matched = [v for k, v in knowledge_dict.items() if k in question]
        return matched if matched else [
            "General guidance: review total spending, refunds, highest categories, and budget status."
        ]


def answer_finance_question_with_rag(user_question):
    """Uses retrieved guidance to answer a finance question."""
    guidance = retrieve_financial_guidance(user_question)
    answer = "Relevant financial guidance:\n"
    for item in guidance:
        answer += f"- {item}\n"
    answer += "\nBudget Buddy recommendation: Use the retrieved guidance together with your transaction analysis to give practical, plain-English advice."
    return answer


# Test
rag_answer = answer_finance_question_with_rag("How should I handle refunds and discretionary spending?")
print(rag_answer)
````

## Custom Financial Tools

````python
# 🤖 AI Collaboration: Custom Savings Calculator Tool
# This tool calculates how long it would take to reach a savings goal.

def savings_goal_calculator(current_savings, monthly_contribution, target_amount):
    """
    Calculates how many months are needed to reach a savings target.
    """
    try:
        current_savings = float(current_savings)
        monthly_contribution = float(monthly_contribution)
        target_amount = float(target_amount)

        if target_amount <= current_savings:
            return "You have already reached your savings goal."

        if monthly_contribution <= 0:
            return "Monthly contribution must be greater than zero."

        remaining_amount = target_amount - current_savings
        months_needed = remaining_amount / monthly_contribution

        return f"You need approximately {months_needed:.1f} months to reach your savings goal."

    except:
        return "Invalid input. Please enter numeric values."


tool_result = savings_goal_calculator(200, 150, 1000)
print(tool_result)
````

````python
# 🤖 AI Collaboration: Multi-Tab Gradio Interface using gr.Blocks
# I had a basic single-tab gr.Interface that worked but looked unprofessional.
# I asked Claude: "Rebuild my Gradio layout as a multi-tab gr.Blocks with three
# tabs — Budget Analysis with CSV upload, AI Chat, and Savings Calculator.
# Use gr.themes.Soft() and wire Enter key submission on the chat tab."
# My improvement: changed the CSV upload label to include column names so
# first-time users immediately know what format to upload. I also fixed the
# chat clear button — Claude's original version didn't include chat_input in
# the outputs list, so the text box wasn't clearing after send.

import gradio as gr
import pandas as pd


def run_budget_analysis(csv_file, weekly_budget):
    if csv_file is None:
        return "⚠️ Please upload a CSV file to begin your analysis."
    try:
        transaction_data = pd.read_csv(csv_file.name)
    except Exception as e:
        return f"❌ Could not read your file. Please check it is a valid CSV.\nError: {str(e)}"
    budget_value = None
    if weekly_budget and str(weekly_budget).strip():
        try:
            budget_value = float(weekly_budget)
        except ValueError:
            return "⚠️ Weekly budget must be a number. Please enter a value such as 120 or 250."
    result = analyze_spending_data(transaction_data, weekly_budget=budget_value)
    if isinstance(result, str):
        return f"❌ {result}"
    return generate_financial_recommendations(result)


def handle_chat_message(user_message, chat_history):
    if not user_message or not user_message.strip():
        return chat_history, ""
    updated_history, _ = budget_buddy_chat(user_message, chat_history)
    return updated_history, ""


def calculate_savings(current_savings, monthly_contribution, savings_goal):
    if current_savings is None or monthly_contribution is None or savings_goal is None:
        return "⚠️ Please fill in all three fields to calculate your savings timeline."
    return savings_goal_calculator(current_savings, monthly_contribution, savings_goal)


with gr.Blocks(theme=gr.themes.Soft(), title="Budget Buddy: Smart Finance Assistant") as demo:

    gr.Markdown(
        """
        # 🏦 Budget Buddy: Smart Finance Assistant
        *Your personal finance companion — spending analysis, AI advice, and savings planning.*
        """
    )

    with gr.Tabs():

        with gr.TabItem("📊 Budget Analysis"):
            gr.Markdown(
                """
                ### Analyse Your Transactions
                Upload your bank transaction CSV and enter your weekly budget to receive
                a personalised spending breakdown and plain-English recommendations.
                """
            )
            with gr.Row():
                with gr.Column(scale=1):
                    csv_upload = gr.File(
                        label="Upload Transactions CSV  (columns: Date, Amount, Category, Description)",
                        file_types=[".csv"]
                    )
                    budget_input = gr.Number(
                        label="Weekly Budget (AUD)",
                        placeholder="e.g. 150",
                        minimum=0
                    )
                    analyse_btn = gr.Button("🔍 Analyse My Spending", variant="primary")
                with gr.Column(scale=2):
                    analysis_output = gr.Textbox(
                        label="Your Spending Report",
                        lines=20,
                        placeholder="Your personalised spending report will appear here.",
                        show_copy_button=True
                    )
            analyse_btn.click(
                fn=run_budget_analysis,
                inputs=[csv_upload, budget_input],
                outputs=analysis_output
            )

        with gr.TabItem("💬 Chat with Budget Buddy"):
            gr.Markdown(
                """
                ### Ask Budget Buddy
                Get practical, non-judgmental financial advice. Ask about budgeting,
                saving, spending habits, or any money question on your mind.
                """
            )
            chatbot = gr.Chatbot(
                label="Budget Buddy",
                height=400,
                placeholder="Hi! I'm Budget Buddy. Ask me anything about your finances. 😊",
                show_copy_button=True
            )
            with gr.Row():
                chat_input = gr.Textbox(
                    label="Your message",
                    placeholder="e.g. How can I spend less on coffee without feeling deprived?",
                    scale=4,
                    lines=1
                )
                send_btn = gr.Button("Send ➤", variant="primary", scale=1)
            clear_btn = gr.Button("🗑️ Clear conversation", variant="secondary")

            send_btn.click(
                fn=handle_chat_message,
                inputs=[chat_input, chatbot],
                outputs=[chatbot, chat_input]
            )
            chat_input.submit(
                fn=handle_chat_message,
                inputs=[chat_input, chatbot],
                outputs=[chatbot, chat_input]
            )
            clear_btn.click(
                fn=lambda: ([], ""),
                outputs=[chatbot, chat_input]
            )
            gr.Examples(
                examples=[
                    ["What is one simple way I can reduce my weekly spending?"],
                    ["How does the 50/30/20 rule work for a student budget?"],
                    ["I spent too much on entertainment this week. What should I do?"],
                    ["How long will it realistically take me to save AUD 2,000?"]
                ],
                inputs=chat_input,
                label="Example questions"
            )

        with gr.TabItem("🎯 Savings Calculator"):
            gr.Markdown(
                """
                ### How Long Until You Reach Your Goal?
                Enter your current savings, your monthly contribution, and your target.
                Budget Buddy will calculate your estimated timeline.
                """
            )
            with gr.Row():
                with gr.Column():
                    current_savings_input = gr.Number(
                        label="Current Savings (AUD)",
                        placeholder="e.g. 500",
                        minimum=0
                    )
                    monthly_contribution_input = gr.Number(
                        label="Monthly Contribution (AUD)",
                        placeholder="e.g. 200",
                        minimum=0
                    )
                    savings_goal_input = gr.Number(
                        label="Savings Goal (AUD)",
                        placeholder="e.g. 3000",
                        minimum=0
                    )
                    calculate_btn = gr.Button("📅 Calculate Timeline", variant="primary")
                with gr.Column():
                    savings_output = gr.Textbox(
                        label="Your Savings Timeline",
                        lines=4,
                        placeholder="Your result will appear here."
                    )
                    gr.Markdown(
                        """
                        **Example scenarios to try:**
                        - Current: AUD 200 | Monthly: AUD 150 | Goal: AUD 1,000 → ~5.3 months
                        - Current: AUD 0 | Monthly: AUD 300 | Goal: AUD 5,000 → ~16.7 months
                        - Current: AUD 2,000 | Monthly: AUD 250 | Goal: AUD 1,500 → already reached!
                        """
                    )
            calculate_btn.click(
                fn=calculate_savings,
                inputs=[current_savings_input, monthly_contribution_input, savings_goal_input],
                outputs=savings_output
            )

    gr.Markdown(
        """
        ---
        *Budget Buddy — ISYS2001 Smart Finance Assistant | Curtin University, Semester 1, 2026*
        """
    )

# demo.launch()
````

## Advanced Integration Tests

````python
import pandas as pd

# 🤖 AI Collaboration: Edge Case Test Suite
# These six tests cover unusual but realistic inputs identified by asking AI
# to review my code for gaps. Edge Test 1 led directly to discovering and
# fixing a real IndexError bug — documented in Developer's Diary Entry 4.

def run_edge_case_tests():
    """
    Tests unusual but realistic scenarios that real users might upload.
    Covers all-refund data, single transactions, whitespace amounts,
    an already-met savings goal, zero contribution, and a budget boundary.
    """
    print("🔍 Running edge case tests...\n")

    # Edge Test 1: All transactions are refunds — no positive spending
    all_refund_data = pd.DataFrame({
        'Date': ['2024-08-01', '2024-08-02'],
        'Amount': ['-$30.00', '-$20.00'],
        'Category': ['Refund', 'Refund'],
        'Description': ['Returned Item 1', 'Returned Item 2']
    })
    result = analyze_spending_data(all_refund_data, weekly_budget=100)
    assert isinstance(result, dict), \
        "Edge Test 1 failed: Should return dict not crash when all transactions are refunds."
    assert result['total_positive_spending'] == 0, \
        "Edge Test 1 failed: Positive spending should be zero."
    assert result['category_summary'].empty, \
        "Edge Test 1 failed: Category summary should be empty."
    print("✅ Edge Test 1 passed: all-refund dataset handled correctly without crashing.")

    # Edge Test 2: Single transaction dataset
    single_data = pd.DataFrame({
        'Date': ['2024-08-01'],
        'Amount': ['$50.00'],
        'Category': ['Groceries'],
        'Description': ['Woolworths']
    })
    result = analyze_spending_data(single_data, weekly_budget=100)
    assert isinstance(result, dict), \
        "Edge Test 2 failed: Single transaction should return dict."
    assert round(result['total_positive_spending'], 2) == 50.00, \
        "Edge Test 2 failed: Total should be 50.00."
    assert len(result['category_summary']) == 1, \
        "Edge Test 2 failed: Should have exactly one category row."
    print("✅ Edge Test 2 passed: single transaction dataset handled correctly.")

    # Edge Test 3: Whitespace-only amount string
    whitespace_data = pd.DataFrame({
        'Date': ['2024-08-01', '2024-08-02'],
        'Amount': ['$40.00', '   '],
        'Category': ['Groceries', 'Shopping'],
        'Description': ['Food', 'Invalid row']
    })
    result = analyze_spending_data(whitespace_data, weekly_budget=100)
    assert round(result['total_positive_spending'], 2) == 40.00, \
        "Edge Test 3 failed: Whitespace row should be dropped."
    print("✅ Edge Test 3 passed: whitespace-only amounts dropped safely.")

    # Edge Test 4: Savings goal already reached
    result = savings_goal_calculator(1000, 150, 500)
    assert "already reached" in result.lower(), \
        "Edge Test 4 failed: Should detect that the savings goal is already met."
    print("✅ Edge Test 4 passed: already-met savings goal detected correctly.")

    # Edge Test 5: Zero monthly contribution
    result = savings_goal_calculator(200, 0, 1000)
    assert "greater than zero" in result.lower(), \
        "Edge Test 5 failed: Zero monthly contribution should be rejected."
    print("✅ Edge Test 5 passed: zero monthly contribution rejected correctly.")

    # Edge Test 6: Spending exactly equal to budget (boundary condition)
    boundary_data = pd.DataFrame({
        'Date': ['2024-08-01'],
        'Amount': ['$100.00'],
        'Category': ['Groceries'],
        'Description': ['Woolworths']
    })
    result = analyze_spending_data(boundary_data, weekly_budget=100)
    assert "within" in result['budget_result'].lower(), \
        "Edge Test 6 failed: Equal spending should show within budget."
    print("✅ Edge Test 6 passed: spending exactly equal to budget correctly shows within budget.")

    print("\n🎉 All 6 edge case tests passed successfully.")
    return "All edge case tests passed successfully."


edge_test_result = run_edge_case_tests()
print(edge_test_result)
````
